```python
counter = 1

   # for i in reviews1:

sentiments = []

for i in reviews1:

    compound_list = []
    positive_list = []
    negative_list = []
    neutral_list = []

#Run Vader Analysis
    compound = analyzer.polarity_scores(i)["compound"]
    pos = analyzer.polarity_scores(i)["pos"]
    neu = analyzer.polarity_scores(i)["neu"]
    neg = analyzer.polarity_scores(i)["neg"]

# Add each value to the appropriate array
    compound_list.append(compound)
    positive_list.append(pos)
    negative_list.append(neg)
    neutral_list.append(neu)

    sentiments.append({"Compound":compound_list,
                       "Positive":positive_list,
                       "Negative":negative_list,
                       "Neutral":neutral_list,
                        "Number Review": counter})

    counter = counter + 1

sentiments_pd = pd.DataFrame.from_dict(sentiments)
sentiments_pd.head()
#print(reviews1)
#Pull in types of food/income/ and group by for x axis.
```

```python
#Create graphs. (swarm w/ violin plot)
#sns.violinplot(x='Number Review', y='Compound', data=sentiments_pd, inner=None)
sns.swarmplot(x='Number Review', y='Compound', data=sentiments_pd)
plt.show()
```
